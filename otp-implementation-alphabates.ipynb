```python
#function to encrypt using OTP
def OTP_encrypt(m, k):
    cipher_msg = ""
    cipher = 0
    for i in range(len(m)):
        cipher=ord(m[i])-ord('A')+ord(k[i])-ord('A')
        if cipher > 25:
            cipher -= 26
        cipher+=65
        cipher_msg = cipher_msg+chr(cipher)
    return cipher_msg    
```

```python
#function to decrypt using OTP
def OTP_decrypt(e, k):
    original_msg = ""
    cipher = 0
    for i in range(len(k)):
        cipher = ord(e[i])-ord('A')-(ord(k[i])-ord('A'))
        if cipher<0:
            cipher+=26
        cipher+=65
        original_msg=original_msg+chr(cipher)
    return original_msg    
```

```python
msg=input("Enter your message without any special characters (in block letters)")
k=input ("Enter the key without any special characters (in block letters)")
if len(msg)!= len(k):
    print("Message length does not match the key length")
    msg=input("Enter your message without any special characters (in block letters)")
    k=input ("Enter the key without any special characters (in block letters)")
if msg.upper()!= msg or k.upper()!= k:
    print("Please provide both the message and key in BLOCK letters")
    msg=input("Enter your message without any special characters (in block letters)")
    k=input ("Enter the key without any special characters (in block letters)")
else :
    s=OTP_encrypt(msg, k)
    print("The encrypted message is:"+s)
    x=OTP_decrypt(s, k)
    print("The decrypted message is:"+x)
```
